    -----Using cosine similarity Algorithm------
    The cosine_similarity function is used to calculate the cosine similarity between pairs of samples. It is commonly used in various fields, including information retrieval, text mining, and recommendation systems, to measure the similarity or distance between vectors.

Here's a brief explanation of the cosine_similarity function and its main functionality:

Cosine similarity: Cosine similarity is a measure of similarity between two non-zero vectors. It calculates the cosine of the angle between the vectors, which ranges from -1 (completely dissimilar) to 1 (completely similar). The closer the cosine similarity value is to 1, the more similar the vectors are.

Pairwise similarity calculation: The cosine_similarity function takes two arrays or matrices as input and calculates the pairwise cosine similarity between the corresponding samples. It computes the cosine similarity for each pair of samples in the input arrays and returns a similarity matrix with dimensions (n_samples_X, n_samples_Y), where n_samples_X and n_samples_Y are the number of samples in the input arrays.

Vector representations: The input arrays or matrices should have the same number of features (dimensions) for each sample. The cosine similarity is calculated based on the vectors' orientations and does not depend on the vectors' magnitudes. Therefore, the vectors are typically normalized before calculating cosine similarity to remove any bias caused by varying vector lengths.

The cosine_similarity function is often used in tasks such as document similarity, recommendation systems, or clustering algorithms that rely on measuring the similarity between samples. It allows for efficient computation of cosine similarity across large datasets and can be a useful tool in various machine learning applications.

The cosine similarity measure is commonly used in various scenarios where the similarity or distance between vectors needs to be assessed. Here are a few common use cases where cosine similarity is employed:

Information retrieval: Cosine similarity is widely used in information retrieval tasks, such as search engines or recommendation systems. In these applications, documents or items are represented as vectors, and cosine similarity is utilized to measure the similarity between a user's query vector and the document or item vectors. This helps in ranking and retrieving the most relevant results.

Text mining and natural language processing (NLP): In NLP tasks, cosine similarity is often employed to compare the similarity between document vectors or text representations. For instance, it can be used to find similar documents, cluster documents based on their topic, or detect plagiarism by comparing the similarity between a given text and a set of reference texts.

Collaborative filtering: In collaborative filtering-based recommendation systems, cosine similarity is used to calculate the similarity between user vectors or item vectors. By measuring the cosine similarity between the vectors representing user preferences or item features, the system can identify similar users or recommend similar items to a given user.

Clustering and dimensionality reduction: Cosine similarity can be utilized as a distance metric in clustering algorithms, such as K-means or hierarchical clustering, when dealing with high-dimensional data. It is particularly useful for text data, where the vectors represent the frequency or TF-IDF values of words. Additionally, cosine similarity is sometimes used in dimensionality reduction techniques, such as Latent Semantic Analysis (LSA) or Latent Dirichlet Allocation (LDA), to identify latent topics or reduce the dimensionality of text data.

Image processing and computer vision: Cosine similarity can also be applied in image processing and computer vision tasks. In these applications, vectors representing image features, such as color histograms or deep learning embeddings, can be compared using cosine similarity to assess their similarity or compute image distances.

These are just a few examples of how cosine similarity is utilized in different domains. The measure's ability to capture the angle between vectors, irrespective of their magnitudes, makes it suitable for assessing similarity or distance in various machine learning and data analysis tasks.

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Data collection and Preprocessing
```

```python
# loading the data from the csv file to apandas dataframe
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.columns
```

```python
movies_data.shape
```

```python
# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director','overview']
print(selected_features)
```

```python
movies_data.isnull().sum()
```

```python
movies_data.info()
```

```python
# replacing the null valuess with null string

for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_features = movies_data[selected_features].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
```

```python
print(combined_features)
```

```python
# Display the first row
print(combined_features.iloc[0])

```

```python
# Display a specific range of rows
print(combined_features.iloc[10:15])
```

```python
print(combined_features.dtype)
print(combined_features.shape)
```

```python
print(combined_features.head())
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#cosine similarity
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

Getting the movie name from the user

```python
# getting the movie name from the user

movie_name = input(' Enter your favourite movie name : ')
```

```python
# creating a list with all the movie names given in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
 #sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key=lambda x: x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

Movie Recommendation Sytem

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
